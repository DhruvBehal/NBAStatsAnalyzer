```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df2223 = pd.read_csv(r"C:\Users\dhruv\Desktop\Data\csv\2223teamstats.csv")
```

```python
print(df2223[['Team', 'ORtg']].head())
```

```python
top5_ortg2223 = df2223.sort_values(by='ORtg', ascending=False)[:5]
```

```python
graph = sns.barplot(x="Team", y="ORtg", data=top5_ortg2223)
graph.set_xticklabels(graph.get_xticklabels(), rotation =45)
graph = graph.set(title="Top 5 Teams Based off ORtgin 22/23")
```
