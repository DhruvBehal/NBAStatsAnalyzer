```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1819 = pd.read_csv(r"C:\Users\dhruv\Desktop\Data\csv\1819teamstats.csv")
print(df1819[['Team', 'DRtg']].head())
top5_drtg1819 = df1819.sort_values(by='DRtg', ascending=True)[:5]
graph = sns.barplot(x="Team", y="DRtg", data=top5_drtg1819)
graph.set_xticklabels(graph.get_xticklabels(), rotation =45)
graph = graph.set(title="Top 5 Teams Based off DRtg in 18/19")
```
