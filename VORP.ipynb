```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1819 = pd.read_csv(r"C:\Users\dhruv\Desktop\Data\csv\1819playerstats.csv")
```

```python
print(df1819[['Player', 'VORP']].head())
```

```python
top5_vorp = df1819.sort_values(by='VORP', ascending=False)[:5]
```

```python
print(top5_vorp)
```

```python
graph = sns.barplot(x="Player", y="VORP", data=top5_vorp)
graph.set_xticklabels(graph.get_xticklabels(), rotation =45)
graph = graph.set(title="Top 5 Players Based off VORP in 18/19")
```
