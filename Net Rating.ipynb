```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1819 = pd.read_csv(r"C:\Users\dhruv\Desktop\Data\csv\1819teamstats.csv")
print(df1819[['Team', 'NRtg']].head())
top5_nrtg1819 = df1819.sort_values(by='NRtg', ascending=False)[:5]
graph = sns.barplot(x="Team", y="NRtg", data=top5_nrtg1819)
graph.set_xticklabels(graph.get_xticklabels(), rotation =45)
graph = graph.set(title="Top 5 Teams Based off NRtg in 18/19")
```
