```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1819 = pd.read_csv(r"C:\Users\dhruv\Desktop\Data\csv\1819playerstats.csv")
print(df1819[['Player', 'USG%']].head())
top5_usg1819 = df1819.sort_values(by='USG%', ascending=False)[:5]
graph = sns.barplot(x="Player", y="USG%", data=top5_usg1819)
graph.set_xticklabels(graph.get_xticklabels(), rotation =45)
graph = graph.set(title="Top 5 Players Based off USG% in 18/19")
```
